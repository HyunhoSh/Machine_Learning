```python
#6-1~6-2
import random

def min_max_avg_med(in_list):
    k = len(in_list)
    in_list.sort()
    min = in_list[0]
    max = in_list[k-1]
    sum = 0
    for i in in_list:
        sum +=i
    avg = sum/k
    avg_ = round(avg,2)
    med = in_list[int(k/2)]
    
    output_list = []
    output_list.append(min)
    output_list.append(max)
    output_list.append(avg_)
    output_list.append(med)
    print("Output:{0}".format(output_list))

s = []
for i in range(9):
    random_int = random.randint(-9,9)
    s.append(random_int)

min_max_avg_med(s)
```

```python
#6-3~6-4
def equ_interval(in_list, n):
    output_list  = []
    
    min = in_list[0]
    max = in_list[1]

    intv = int((max-min)/n)

    nlist = []
    
    
    tmp = min

    for i in range(n):
        nlist.append(tmp)
        nlist.append(tmp+intv)
        tmp = tmp+intv

        output_list.append(nlist)
        
        nlist = []
        

    print("Output:{0}".format(output_list))

equ_interval([-4,8],3)
equ_interval([-2,4],4)
```

```python
#7
#7-1
import matplotlib.pyplot as plt
import csv

def no_of_class_values(in_list, c):
    
    r = len(in_list)
    line  = []
    
    for i in range(r):
        if in_list[i][-1] == c:
            line.append(in_list[i][:-1])
            
    if line == []:
        print("None")
    else:
        return line
```

```python
#7-2        
def no_of_dis_val(in_list):
    
    k = list(set(in_list))
    l = len(k)
    return l
```

```python
#7-3
f= open('C:\\\\\\dataset_1\\audiology.csv', 'r', encoding = 'utf-8')
rdr = csv.reader(f)
a_list = []

next(rdr) 

for line in rdr:
    a_list.append(line)

f.close()

class_data = [] 
for list_ in a_list:
    class_data.append(list_[-1])

class_data_ = list(set(class_data))

line_1 = []
line_2 = []
for i in class_data_:
    s = no_of_class_values(a_list, i)
    line  = sum(s, [])
    m = no_of_dis_val(line)
    line_1.append(i)
    line_2.append(m)
    print("{0}: {1}".format(i, m))
```

```python
#7-4
ax = plt.subplot(1,1,1)
ax.bar(line_1, line_2, color= 'Blue', width = 0.5)
plt.xlabel('Name of the class')
plt.ylabel = ('The number of values')
plt.xticks(rotation='90')
# for label in ax.xaxis.get_ticklabels():
#     label.set_rotation(45)

plt.grid()
plt.show()
```
